```python
import pandas as pd
```

# Inicio

```python
pessoas = pd.read_excel('casodeestudo.xlsx', sheet_name='pessoas')
covid = pd.read_excel('casodeestudo.xlsx', sheet_name='registros_covid')
pessoas
```

```python
pessoas.groupby('sexo').idade.mean() #Idade media de dexo feminino e masculino
```

```python
pessoas['inicial'] = pessoas.nome.str[0] # Criando uma nova coluna 'inicial' 
pessoas
```

```python
pessoas.groupby('inicial').idade.mean() # Pegando o agupamento pela coluna 'inicial', pegando apenas a idade e aplicando a media a ela.
```

# Join

```python
pessoas.head()
```

```python
covid.head()
```

```python
pessoas = pessoas.set_index('id') # Com isso estou falando que a coluna 'id' deve ser o index para tabela pesoas
covid = covid.set_index('id') # Com isso estou falando que a coluna 'id' deve ser o index para tabela covid
```

```python
pessoas.head(5)
```

```python
covid.head(5)
```

```python
df = pessoas.join(covid.set_index('id_pessoa')) # Unificando a tablela pessoas com a tablela covid e a coluna id_pessoa seja index
df
```

```python
df = df.fillna(0) # Substituindo os NaNs da coluna severidade. Esses NaN foram gerados pelo join. Não é recomandado que uma tabela possua dados NaN
df
```

```python
df.groupby('severidade').idade.mean() # Pegando a media da coluna severidade.
```

# Novas Features

```python
df['faixa_etaria'] = 'Jovem'
df.head(5)
```

```python
df.loc[df.idade>24,'faixa_etaria'] = 'adulto'
df.loc[df.idade>45,'faixa_etaria'] = 'adulto2'
df.head()
```

```python
df.loc[df.severidade=='baixa', 'severidade'] = 1 # Passa os dados da coluna severidade para numeros do tipo string
df.loc[df.severidade=='alta', 'severidade'] = 2
df.head(10)
```

```python
df['severidade'] = df['severidade'].astype('int') # Passa os dados da coluna severidade para tipo numerios inteiros
```

```python
df.groupby('faixa_etaria').mean()
```
